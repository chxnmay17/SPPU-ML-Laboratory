```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Define the Function and its Derivative ---

def f(x):
    """The function we want to minimize."""
    return (x + 3)**2

def f_derivative(x):
    """The derivative (gradient) of the function."""
    # The derivative of (x+3)^2 is 2*(x+3)
    return 2 * (x + 3)

# --- 2. Set Parameters for Gradient Descent ---

# The algorithm starts at this x-value
current_x = 2.0

# The "step size". A larger value takes bigger steps,
# a smaller value takes smaller (and safer) steps.
learning_rate = 0.1

# The number of steps to take.
iterations = 100

# We will store the history of x and y values to plot them later
x_history = [current_x]
y_history = [f(current_x)]

# --- 3. Run the Gradient Descent Algorithm ---

print(f"Starting Gradient Descent to find the minima of y = (x+3)^2")
print(f"Starting at x = {current_x}\n")

for i in range(iterations):
    # Calculate the gradient (the slope) at the current x
    gradient = f_derivative(current_x)

    # Update the x value by taking a small step in the
    # *opposite* direction of the gradient.
    # This is the core formula: x_new = x_old - (learning_rate * gradient)
    current_x = current_x - (learning_rate * gradient)

    # Store the new x and y values for our plot
    x_history.append(current_x)
    y_history.append(f(current_x))

    # Print the progress every 10 steps
    if (i+1) % 10 == 0:
        print(f"Iteration {i+1:3}: x = {current_x: .5f}, y = f(x) = {f(current_x): .5f}")

# --- 4. Print the Final Result ---
print("\n--- Gradient Descent Finished ---")
print(f"The algorithm converged to a local minima at:")
print(f"x ≈ {current_x: .6f}")
print(f"y ≈ {f(current_x): .6f}")

# The true minima is at x = -3, so the algorithm should be very close.

# --- 5. Plot the Convergence ---
print("\nPlotting the convergence...")

# Create x-values for the original function plot
plot_x = np.linspace(-9, 3, 100)
# Create y-values for the original function plot
plot_y = f(plot_x)

plt.figure(figsize=(10, 6))
# Plot the function y=(x+3)^2
plt.plot(plot_x, plot_y, label="Function: y = (x+3)²", zorder=1)

# Plot the path of the gradient descent
plt.plot(x_history, y_history, 'o-', c='red', label="Path of Gradient Descent", zorder=2)

# Mark the starting point
plt.scatter(x_history[0], y_history[0], c='green', s=100, zorder=3, label="Start Point (x=2)")

# Mark the final point
plt.scatter(x_history[-1], y_history[-1], c='black', s=100, zorder=3, label="Final Minima")

plt.title("Gradient Descent Visualization")
plt.xlabel("x")
plt.ylabel("y = f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
